**1.2 Importing the modules**

```python
import keras_ocr
import matplotlib.pyplot as plt
```

**1.3 Downloading the weights of ICR**

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

**1.4 Calling the images**

```python
images = [keras_ocr.tools.read(img) for img in ['https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSr3fbCM8TdYiAvYwpf-csQ5ny03h7TS1jlag&usqp=CAU','https://static.toiimg.com/photo/82940202.cms']]
```

**1.5 Displaying the image**

```python
plt.figure(figsize = (10,20))
print(plt.imshow(images[0]))
```

```python
plt.figure(figsize = (10,20))
print(plt.imshow(images[1]))
```

**1.6 Prediction and Result**

```python
prediction_groups = pipeline.recognize(images)
fig, axs = plt.subplots(nrows=len(images), figsize=(10, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, 
                                    predictions=predictions, 
                                    ax=ax)
```
